```python
# can open a file in 2 ways

file = open("./sample_data/sample.txt", "r")
content = file.read()
print(content)
print(type(content))
```

```python
file = open("./sample_data/sample.txt", "r")
content = file.readlines()
print(content)
print(file.closed)
file.close()
print(file.closed)
```

```python
with open("./sample_data/sample.txt", "r") as file:
  content = file.readlines()
  print(content)
print(file.closed)
```

```python
with open("./sample_data/sample.txt", "r") as file:
  for line in file:
    # print(line, end="")
    l = line.strip()
    print(l)
```

```python
# write in a file by creating

with open("./sample_data/test.txt", "w") as file:
  file.write("Hello test file\n")
  file.write("Ai ml batch 02\n")
  file.write("learning file read write")

print(file.closed)
```

```python
# writing on a existing file

with open("./sample_data/test.txt", "w") as file:
  file.write("second attemp\n")
  file.write("Ai ml batch 02\n")


```

```python
# write or appending in existing file without over writting
with open("./sample_data/test.txt", "a") as file:
  file.write("second attemp\n")
  file.write("Ai ml batch 02")
  file.write("cse sucks, totally sucks")


```

```python
with open("./sample_data/test.txt") as file:
  content = file.readlines()
  content = list(map(str.strip, content))

  filter_content = list(filter(lambda x : len(x) > 13, content))
print(filter_content)
```
