```python
numbers = input().split()
even_sum = 0
even_count = 0

for num in numbers:
    num = int(num)

    if num % 2 == 0:
        even_sum += num
        even_count += 1

print(even_count, even_sum)
```

```python
# generator
import numpy as np

data = np.arange(20)
# print(data)


def data_loader(batch_size, data):
    for i in range(0, len(data), batch_size):
        yield data[i : i + batch_size]


loader = data_loader(5, data)
print(next(loader))
print(next(loader))
print(next(loader))
```
