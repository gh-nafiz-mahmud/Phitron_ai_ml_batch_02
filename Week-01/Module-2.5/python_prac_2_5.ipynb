```python
# problem 2
"""
Note: Take help from support session to solved this problem (mainly yield data[.......] session)
"""
data = [1, 2, 3, 4, 5, 6, 7, 8]
batch_size = 3


def batch_generator(data, batch_size):
    for i in range(0, len(data), batch_size):
        yield data[i : i + batch_size]


ans = batch_generator(data, 3)
print(next(ans))
print(next(ans))
print(next(ans))
# print(next(ans))
```
