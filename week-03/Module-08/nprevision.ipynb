```python
dick = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}
print(dick)
print(type(dick))

array_dick1 = np.array(list(dick.keys()))
array_dick2 = np.array(list(dick.values()))
array_dick3 = np.array(list(dick.items()))

print(array_dick1)
print(array_dick2)
print(array_dick3)

print(array_dick1.ndim)
print(array_dick2.ndim)
print(array_dick3.ndim)

print(array_dick1.size)
print(array_dick2.size)
print(array_dick3.size)

print(array_dick3.dtype)
print(array_dick2.dtype)
print(array_dick1.dtype)
```

```python
arr = np.array([[1, 2], [3, 4]])
print(arr.ndim)
flat1 = arr.flatten()
flat2 = arr.ravel()
print("Flatten result:", flat1, type(flat1), flat1.ndim, flat1.dtype)
print("Ravel result:", flat2, type(flat2), flat2.ndim, flat2.dtype)
```

```python
arr = np.array([[10, 20], [30, 40]])
r = arr.flatten()
r[0] = 999
print("Modified flatten array:", r)
print("Original array also changes:", arr)
```

```python
arr = np.array([[10, 20], [30, 40]])
r = arr.ravel()
r[0] = 999
print("Modified ravel array:", r)
print("Original array also changes:", arr)
```

```python
arr = np.array([1, 2, 3])
print(arr.shape)

expanded = np.expand_dims(arr, axis=0)
print(expanded.shape)

arr = np.squeeze(arr)
print(arr.shape)

expanded1 = np.expand_dims(arr, axis=1)
print(expanded1.shape)

arr = np.squeeze(arr)
print(arr.shape)
```
