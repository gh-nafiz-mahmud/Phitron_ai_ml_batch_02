```python
matrix = [
    [1,2,3],
    [4,5,6]
]

arr_matrix = np.array(matrix, dtype=np.float32)
print(matrix)
print(arr_matrix)

print(type(matrix))
print(arr_matrix.ndim)
print(arr_matrix.shape)

```

```python
st = {1,2,3}
st_array = np.array(list(st))

print(st_array.dtype)
print(st_array.ndim)
print(st_array.shape)
print(st_array.size)

```

```python
dick = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}
print(dick)
print(type(dick))

array_dick1 = np.array(list(dick.keys()))
array_dick2 = np.array(list(dick.values()))
array_dick3 = np.array(list(dick.items()))

print(array_dick1)
print(array_dick2)
print(array_dick3)

print(array_dick1.ndim)
print(array_dick2.ndim)
print(array_dick3.ndim)

print(array_dick1.size)
print(array_dick2.size)
print(array_dick3.size)

print(array_dick3.dtype)
print(array_dick2.dtype)
print(array_dick1.dtype)
```

```python
arr = np.zeros((3,3))
arr = arr.astype(np.int16)

print(arr)
print(arr.dtype)

arr2 = np.zeros_like(arr)
print(arr2)
print(arr2.dtype)
```

```python
arr = np.ones((3,3), dtype=np.int16)
print(arr)

arr2 = np.ones_like(arr)
print(arr2)
print(arr2.dtype)
```

```python
arr = np.empty((3,3), dtype=np.int8)
print(arr)

arr2 = np.empty_like(arr)
print(arr2)
print(arr2.dtype)
```

```python
arr = np.full((3,3), fill_value=1)
arr2 = np.full((3,3), np.inf)

print(arr)
print(arr2)
```

```python
arr = np.full((3,3), np.inf)
print(arr)
arr2 = np.full_like(arr, np.inf, dtype=np.float32)
print(arr2)
```

```python
arr = np.arange(1,17)
print(arr)

reshape_arr = arr.reshape(4,4)
print(reshape_arr)
```

```python
arr = np.linspace(1,2,5, dtype=np.float32)
print(arr)
```
