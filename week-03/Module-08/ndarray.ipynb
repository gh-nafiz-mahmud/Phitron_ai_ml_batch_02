```python
import numpy as np
```

```python
# 1d array

arr1 = np.array([1,2,3,4,5])
print(arr1)
print(type(arr1))
print(arr1.ndim)


```

```python
# 2d array

arr2 = np.array([[1,2,3], # row 1
                 [4,5,6] # row 2
                 ])

print(arr2)
print(arr2.ndim)
```

```python
# 3d array

arr3 = np.array([
    [1,2,3],
    [4,5,6],
])

print(arr3)
print(arr3.ndim)
```
